```python
import pandas as pd

# Load the dataset
bank_data = pd.read_excel("C:/Users/Horizon/Downloads/Banks Data.xlsx")

# Display the original DataFrame
print("Original DataFrame:")
print(bank_data.head())

# Separate the 'Name' column into 'Bank' and 'Variable'
bank_data[['Bank', 'Variable']] = bank_data['Name'].str.extract(r'^\s*(.+?)\s*-\s*(.+?)\s*$')
bank_data = bank_data.drop(columns=['Name'])

# Melt the DataFrame to long format
melted_data = pd.melt(bank_data, id_vars=['Bank', 'Variable'], var_name='Year', value_name='Value')

# Pivot the DataFrame to get the desired tidy format
tidy_bank_data = melted_data.pivot_table(index=['Bank', 'Year'], columns='Variable', values='Value', aggfunc='first')

# Reset the index to turn multi-index into columns
tidy_bank_data = tidy_bank_data.reset_index()

# Display the tidy DataFrame
print("Tidy DataFrame:")
print(tidy_bank_data.head())
```

```python
import pandas as pd

# Load the dataset
bank_data =pd.read_excel("C:/Users/Horizon/Downloads/Banks Data.xlsx")
# Separate the 'Name' column into 'Bank' and 'Variable'
bank_data[['Bank', 'Variable']] = bank_data['Name'].str.extract(r'^\s*(.+?)\s*-\s*(.+?)\s*$')

# Drop the original 'Name' column
bank_data = bank_data.drop(columns=['Name'])

# Melt the DataFrame to long format
melted_data = pd.melt(bank_data, id_vars=['Bank', 'Variable'], var_name='Year', value_name='Value')

# Drop rows with NaN values in 'Value'
melted_data = melted_data.dropna(subset=['Value'])

# Pivot the DataFrame to get the desired tidy format
tidy_bank_data = melted_data.pivot_table(index=['Bank', 'Year'], columns='Variable', values='Value', aggfunc='first')

# Reset the index to turn multi-index into columns
tidy_bank_data = tidy_bank_data.reset_index()

# Fill NaN values with a specified value, e.g., 0
tidy_bank_data = tidy_bank_data.fillna(0)

# Display the tidy DataFrame
print("\nTidy DataFrame:")
print(tidy_bank_data.head())
```
